```python
def Lk(x,k,xp):
    p = 1
    for i in range(len(xp)):
        if i != k:
            p = p * (x - xp[i])/(xp[k] - xp[i])
    
    return p 
    
def pk(x,xp,yp):
    sum = 0  
    for idx, y in enumerate(yp):
        sum += y * Lk(x,idx,xp) 
    
    return sum 
    
```

```python
x = np.linspace(0, np.pi, 5)
y = [np.sin(x) for x in x] 
print(x, y)

for i in range(5):
    print("y:", pk(x[i], x, y))
```
